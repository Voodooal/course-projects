### Problem 1: More Fun with Tuples

```python
# 1.1
def my_sum(*args):
    total = 0
    for n in args:
        total = total + n
    return total
```

```python
# 1.2
def reverse_tuple(t):
    lst = list(t)
    for i in range(len(t)):
        lst[i] = t[-(i+1)]
    return tuple(lst)
```

```python
# 1.3
def rotate_tuple(t, n):
    lst = list(t)
    try:
        if not isinstance(t, tuple): raise TypeError
        if not isinstance(n, int): raise TypeError
        for i in range(len(t)):
            if n > 0:
                lst[i] = t[i-n]
            elif n < 0:
                lst[i+n] = t[i]
            else:
                lst = t
        return tuple(lst)
    except TypeError: 
        if not isinstance(t, tuple):
            print("The first input should be a tuple.")
        elif not isinstance(n, int):
            print("The second input should be an integer. It will be casted to an integer.")
            return rotate_tuple(t, int(n))
```

### Problem 2: More Fun with Vectors
Problem 2 took me 30 minutes. 

```python
# 2.1
def is_valid_sparse_vector(v):
    if not isinstance(v, dict):
        return False
    for key in v:
        if not (isinstance(key, int) and key > 0):
            return False    
        if not isinstance(v[key], float):
            return False
    return True
```

```python
# 2.2
def sparse_inner_product(v1, v2):
    if not(is_valid_sparse_vector(v1) and is_valid_sparse_vector(v2)):
        raise TypeError("The input is not a proper vector.")
    prod = 0
    for i in v1:
        for j in v2:
            if i == j:
                prod = prod + v1[i] * v2[j]
    return prod
```

### Problem 3: Counting Word Bigrams
Problem 3 took me 4 hours. 

In this problem, I split each line of the file by whitespace and then strip each word with string punctuation and badstrings. I remove all the empty strings from the word list. By doing this, words or phrases concatenated with the punctuations will not be split and remain as a whole string. 

```python
# 3.1
import string
with open('badstring.txt') as f:
    badstring = f.read()
f.close()

def count_bigrams_in_file(filename):
    try:
        if not isinstance(filename,str): raise TypeError
        wd_lst = []
        with open(filename) as f:
            for line in f:
                for wd in line.split():
                    wd_lst.append(wd.lower().strip().strip(string.punctuation + badstring))
        wd_lst = [i for i in wd_lst if i != ""]
        
        bigram_dict = dict()
        for i in range(len(wd_lst)-1):
            key_tuple = (wd_lst[i], wd_lst[i+1])
            bigram_dict[key_tuple] = bigram_dict.get(key_tuple, 0) + 1
        return bigram_dict
    except TypeError:
        print("The input should be a string.")
    except:
        print("Cannot load the file.")
```

```python
# 3.2
import pickle
mb_bigram = count_bigrams_in_file("mobydick.txt")
#pickle.dump(mb_bigram, open('mb.bigram.pickle','wb'))
pickle.load(open('mb.bigram.pickle','rb'))
```

```python
# 3.3
import string
with open('badstring.txt') as f:
    badstring = f.read()
f.close()

def collocations(filename):

    wd_lst = []
    with open(filename) as f:
        for line in f:
            for wd in line.split():
                wd_lst.append(wd.lower().strip().strip(string.punctuation + badstring))
    wd_lst = [i for i in wd_lst if i != ""]
        
    colloc_dict = dict()
        
    n = len(wd_lst)
    if n == 1:
        colloc_dict[wd_lst[0]] = set()
    else:
        colloc_dict[wd_lst[0]] = {wd_lst[1], }
        colloc_dict[wd_lst[n-1]] = {wd_lst[n-2], }
        for i in range(1, n-1):
            if wd_lst[i] not in colloc_dict.keys():
                colloc_dict[wd_lst[i]] = {wd_lst[i-1], wd_lst[i+1]}
            else:
                colloc_dict[wd_lst[i]].add(wd_lst[i-1])
                colloc_dict[wd_lst[i]].add(wd_lst[i+1])
        
    return colloc_dict
  
```

```python
# 3.4
mb_colloc = collocations("mobydick.txt")
#pickle.dump(mb_colloc, open('mb.colloc.pickle','wb'))
pickle.load(open('mb.colloc.pickle','rb'))
```
